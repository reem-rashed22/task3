```python
import networkx as nx
```

```python
graph= nx.Graph()
graph.add_nodes_from(['x', 'y', 'z', 'm'])
graph.add_edges_from([('x','y'), ('y','z'), ('z','m'), ('x','z')])
nx.draw(graph, with_labels = True)
```

```python
list(graph.neighbors('y'))
```

```python
for neighbor in graph.neighbors('y'):
    print(neighbor)
```

```python
graph.has_node('x')
```

```python
nx.is_tree(graph)
```

```python
nx.is_connected(graph)
```

```python
graph.degree('x')
```

```python
draw= nx.DiGraph()
draw.add_edges_from([('10','20'), ('20','30'), ('10','40'), ('30','40')])
nx.draw(draw, with_labels = True)
```

```python
G=nx.Graph()
G.add_edges_from([(10,20),(20,30),(10,30),(30,40)])
nx.draw(G,with_labels=True,node_size=500)
```

```python
dict(G.degree())
```

```python
nx.to_numpy_matrix(G)
```
